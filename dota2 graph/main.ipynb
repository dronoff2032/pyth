Цель работы: научится работать с matplotlib и проанализировать датасет Dota2 Heroes by MMR Bracket.

Здесь производится иvпорт нужных библиотек и чтение файлов

```python
import matplotlib.pyplot as plt
import pandas as dp
table = dp.read_csv('Dota2 Heroes by MMR Bracket.csv')
```

```python
x = table['Heroes'].tail(10)
y = [float(percent.strip('%')) for percent in table['(<2K MMR) PickRate(%)'].tail(10)]

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Процент пика')
plt.bar(x, y, label='Процент пика')
plt.title('Статистика пика персонажей меньше 2к ммр')
plt.xlabel('Герой')
plt.ylabel('Процент пика')
plt.legend()
plt.show()
```

```python
x = table['Heroes'].tail(10)
y = [float(percent.strip('%')) for percent in table['(2K-3K MMR) PickRate(%)'].tail(10)]

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Процент пика')
plt.bar(x, y, label='Процент пика')
plt.title('Статистика пика персонажей от 2к до 3к ммр')
plt.xlabel('Герой')
plt.ylabel('Процент пика')
plt.legend()
plt.show()
```

```python
x = table['Heroes'].tail(10)
y = [float(percent.strip('%')) for percent in table['(3K-4K MMR) PickRate(%)'].tail(10)]

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Процент пика')
plt.bar(x, y, label='Процент пика')
plt.title('Статистика пика персонажей от 3к до 4к ммр')
plt.xlabel('Герой')
plt.ylabel('Процент пика')
plt.legend()
plt.show()
```

Исходя из этих 3 графиков можно сделать вывод, что при повышении опыта игры этих персонажей не перестали выбирать реже, за исключением Hoodwink, популярность которого из графика в график снижалась.

```python
x = [float(percent.strip('%')) for percent in table['(<2K MMR) PickRate(%)'].tail(10)]
x_avr = sum(x) / len(x)
y = [float(percent.strip('%')) for percent in table['(<2K MMR) WinRate(%)'].tail(10)]
y_avr = sum(y) / len(y)
vals = [x_avr, y_avr]
plt.pie(vals, labels=['Средний пик рейт', 'Средний вин рейт'])
plt.title('Соотношение среднего пик рейта и винрейта до 2к ммр')
plt.show
```

```python
x = [float(percent.strip('%')) for percent in table['(3K-4K MMR) PickRate(%)'].tail(10)]
x_avr = sum(x) / len(x)
y = [float(percent.strip('%')) for percent in table['(3K-4K MMR) WinRate(%)'].tail(10)]
y_avr = sum(y) / len(y)
vals = [x_avr, y_avr]
plt.pie(vals, labels=['Средний пик рейт', 'Средний вин рейт'])
plt.title('Соотношение среднего пик рейта и винрейта от 3к до 4к ммр')
plt.show
```

Исходя из этих 2 круговых диаграмм можно сделать вывод, что соотношение между средним пик рейтом и средним вин рейтом не меняется значительно от получения опыта игры. Зависимости не обнаружено.

```python
x = [float(percent.strip('%')) for percent in table['(2K-3K MMR) PickRate(%)'].tail(20)]
y = [float(percent.strip('%')) for percent in table['(2K-3K MMR) WinRate(%)'].tail(20)]

plt.figure(figsize=(5, 5))
plt.scatter(x, y)
plt.title('Статистика пика персонажей от 2к до 3к ммр')
plt.xlabel('Процет вин рейта')
plt.ylabel('Процент пик рейта')
plt.legend()
plt.show()
```

```python
x = [float(percent.strip('%')) for percent in table['(>5K MMR) PickRate(%)'].tail(20)]
y = [float(percent.strip('%')) for percent in table['(>5K MMR) WinRate(%)'].tail(20)]

plt.figure(figsize=(5, 5))
plt.scatter(x, y)
plt.title('Статистика пика персонажей от 5к и больше ммр')
plt.xlabel('Процет вин рейта')
plt.ylabel('Процент пик рейта')
plt.legend()
plt.show()
```

Исходя из этих 2 рассеяных гистограм можно сделать вывод, что в последствии увеличения опыта в игре, для тех героев, у кого и так было низкое соотношение пик рейти и вин рейта стало еще меньше, а у тех, у кого это соотношение было большим стало еще больше. Усредненных значений стало меньше всего.

```python
data = [float(percent.strip('%')) for percent in table['(<2K MMR) PickRate(%)'].tail(125)]
plt.hist(data, bins=50)
plt.title('Статистика пика персонажей до 2к ммр')
plt.legend()
plt.show

```

```python
data = [float(percent.strip('%')) for percent in table['(3K-4K MMR) PickRate(%)'].tail(125)]
plt.hist(data, bins=50)
plt.title('Статистика пика персонажей от 3к до 4к ммр')
plt.legend()
plt.show
```

```python
data = [float(percent.strip('%')) for percent in table['(>5K MMR) PickRate(%)'].tail(125)]
plt.hist(data, bins=50)
plt.title('Статистика пика персонажей от 5к и больше ммр')
plt.legend()
plt.show
```

Исходя из этих гистограмм была подтверждена тенденция на укрепление позиций часто выбираемых персонажей и на падение редко выбираемых.
